```python
#dependencies
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from palettable.colorbrewer.sequential import Reds_9, Blues_9, Purples_9
from PIL import Image
from wordcloud import WordCloud, STOPWORDS
```

```python
#make the functions for the color palettes, leaving out the whitish colors
def blue(word, font_size, position, orientation, random_state=None, **kwargs):
    return tuple(Blues_9.colors[random.randint(2,8)])

def red(word, font_size, position, orientation, random_state=None, **kwargs):
    return tuple(Reds_9.colors[random.randint(2,8)])

def purple(word, font_size, position, orientation, random_state=None, **kwargs):
    return tuple(Purples_9.colors[random.randint(2,8)])
```

```python
#set up the site
site = 'combined'

#pick a side: 'R', 'L', or 'All'
side = 'all'

#pick a color: red, blue, or purple
color = purple

#pick a shape: 'donkey', 'elephant', or 'usa'
shape = 'usa'
```

```python
#read in the news site's text
text = ''
with open(f'{side}-{site}-text.txt', encoding='utf-8') as file:
    for line in file:
        text += line
```

```python
print(len(text))
print(text[0:50])
```

```python
text_list = text.split(' ')
df = pd.DataFrame({'text': text_list})
```

```python
print(df.shape)
```

```python
#set up stopwords
stopwords = set(STOPWORDS)

#read in the extra stopwords
extra_stopwords = open('../resources/extra_stopwords.txt').read().split()

#add the extra stopwords
for word in extra_stopwords:
    stopwords.add(word)
    
#add special cases to stopwords
stopwords.update({'Washington Post'})
```

```python
#read in the mask
mask_png = f'../resources/{shape}.png'
mask = np.array(Image.open(mask_png))
```

```python
#set up the wordcloud
wc = WordCloud(background_color="white", width = 800, height = 400, max_words=2600, mask=mask, 
               color_func=color, stopwords=stopwords, random_state=13)
```

```python
# generate word cloud
wc.generate(text)
```

```python
#run to recolor the words if desired
wc.recolor(color_func=color, random_state=3)
```

```python
#plot
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
#export png
wc.to_file(f'../output/img/{side}-{site}.png')
```
