```python
#dependencies
import pymongo
import time
```

```python
#set up mongodb connection
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client['FINALP']
```

```python
#list of news sites
name_list = ['BBC', 'americanconservative', 'atlantic', 'breitbart', 'dailywire', 'economist', 'fiscaltimes', 
             'foxnews', 'mjones', 'newrepublic', 'nypost', 'politico', 'reason', 'slate', 'thedailybeast', 
             'theguardian', 'thehill', 'theintercept', 'washpost', 'washtimes']

#for collection_name in name_list:
#    documents = db[collection_name].find()
#    print(documents.count())
```

```python
# Record starting time
start = time.time()

#blank list
text_list = []

#loop through the sites
for name in name_list:
    
    #chose the HTML_pages
    documents = db[name].find()
    
    print()
    print(name)
    print('-----------------------')
    
    #loop through the documents
    for i in range(0, documents.count()):
    
        #progress tracker
        print(f'{i}/{documents.count()}')
        
        text = (documents[i]['meta']['text']).replace("\n", " ")
                       
        #append to text_list
        text_list.extend(text.split(' '))
    
#end time
end = time.time()

#minutes elapsed
print((end - start)/60)    
```
