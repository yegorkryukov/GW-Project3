```python
#dependencies
import pandas as pd
```

```python
#read in csvs as dfs
start_site = 'WashingtonTimes'
media_names = ['NewYorkPost', 'Hill', 'FoxNews', 'FiscalTimes', 
         'Economist', 'DailyWire','AmericanConservative', 
         'Breitbart', 'Reason']

right_country_df = pd.read_csv(f'../output/R-{start_site}3_country.csv', index_col=0)
#drop the percentage column
right_country_df.drop(['Percentage'], axis=1, inplace=True)

for name in media_names:
    df = pd.read_csv(f'../output/R-{name}3_country.csv', index_col=0)
    #drop the percentage column
    df.drop(['Percentage'], axis=1, inplace=True)
    #add the df to the country_df
    right_country_df = right_country_df.add(df, fill_value=0)

#sort descending
right_country_df.sort_values(by=['Count'], ascending=False, inplace=True)
right_country_df
```

```python
#Get the count total
country_total = right_country_df.sum(axis=0)

#pull out the actual number
country_total = country_total.iloc[0]

country_total
```

```python
#get the percentages and make a column
right_country_df['Percentage'] = [(row/country_total) * 100 for row in right_country_df['Count']]

right_country_df.head(10)
```

```python
#drop the count column
right_country_df = right_country_df[['Percentage']]

right_country_df.head()
```

```python
#export as csv
right_country_df.to_csv('../output/Right_country.csv')
```
